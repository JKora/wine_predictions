```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_csv("filtered_wine_data.csv")
df.head()
```

```python
stop = stopwords.words('english')
df["description"] = df["description"].str.lower().str.split()
```

```python
df['filtered'] = df['description'].apply(lambda x: ' '.join([word for word in x if word not in (stop)]))
df.head()
```

```python
tokenized_sents = [word_tokenize(i) for i in df.filtered]
```

```python
tokenized_sents
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokenized_sent=[]
for row in df.filtered:
    tokenized_sent.append(tokenizer.tokenize(row))
```

```python
df['tokenized_sent'] = tokenized_sent
```

```python
df.head()
```

```python
df.to_csv("tokenized_wine_data.csv")
```
