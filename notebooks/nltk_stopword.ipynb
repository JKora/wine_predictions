```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df_variety = pd.read_csv("filtered_wine_data.csv")
df_variety.head()
```

```python
df_country = pd.read_csv("filtered_wine_data_country.csv")
df_country.head()
```

```python
df_winery = pd.read_csv("filtered_wine_data_winery.csv")
df_winery.head()
```

```python
stop = stopwords.words('english')
df_variety["description"] = df_variety["description"].str.lower().str.split()
df_country["description"] = df_country["description"].str.lower().str.split()
df_winery["description"] = df_winery["description"].str.lower().str.split()
```

```python
df_variety['filtered'] = df_variety['description'].apply(lambda x: ' '.join([word for word in x if word not in (stop)]))
df_variety.head()
```

```python
df_country['filtered'] = df_country['description'].apply(lambda x: ' '.join([word for word in x if word not in (stop)]))
df_country.head()
```

```python
df_winery['filtered'] = df_winery['description'].apply(lambda x: ' '.join([word for word in x if word not in (stop)]))
df_winery.head()
```

```python
#tokenized_sents = [word_tokenize(i) for i in df.filtered]
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
variety_tokenized_sent=[]
for row in df_variety.filtered:
    variety_tokenized_sent.append(tokenizer.tokenize(row))
```

```python
df_variety['tokenized_sent'] = variety_tokenized_sent
```

```python
df_variety.head()
```

```python
df_variety.to_csv("tokenized_wine_data.csv")
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
country_tokenized_sent=[]
for row in df_country.filtered:
    country_tokenized_sent.append(tokenizer.tokenize(row))
```

```python
df_country['tokenized_sent'] = country_tokenized_sent
```

```python
df_country.head()
```

```python
df_country.to_csv("tokenized_wine_data_country.csv")
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
winery_tokenized_sent=[]
for row in df_winery.filtered:
    winery_tokenized_sent.append(tokenizer.tokenize(row))
```

```python
df_winery['tokenized_sent'] = winery_tokenized_sent
```
